The blockchain is a distributed ledger the serves as a verifiable record of financial transactions. It works by creating 'links' in a cryptographic 'chain' of computations. Each 'link' adds a set of transactions to the chain. The algorithm for adding new links$^1$ is designed to be costly both in terms of electricity and time. This is an example where an algorithm is designed to be slow and inefficient. By making the algorithm slow, the blockchain can be secured against attackers without trusting a central authority (like a company or government) to manage the blockchain. The amount of time to generate any given link is random, but over time the global network reaches an average rate of generating links (through the law of large numbers).

Each new link requires all of the previous links. Links must be generated in sequential order, one at a time. In order to falsify an alternate version of a recent link in the blockchain (e.g., to steal bitcoins), an attacker would need to recompute every link between the falsified link and the newest link in the chain. However, the global blockchain network is continuously generating links at a constant rate $r_{GN}>0$. Thus, in order to create a alternate version of the blockchain ledger, the attacker would need to compute links sufficiently fast to overtake the global network.

Assuming that the attacker cannot obtain sufficient computational power to generate links at higher average rate than the global network, the only way to alter the ledger is to get lucky. The attacker would need to have a streak of faster than average links that overtakes the global network. Assuming that the attacker can generate links at an average rate $0 < r_A < r_{GN}$, the probability of achieving a lucky streak is the solution to the following differential equation. **Note that this is a qualitatively accurate idealization.**

Let $r = r_{GN} - r_A = 3.5$. Let $x$ be the initial distance (in blockchain links) from the most current link generated by the global network. Let us assume that the attacker will give up after reaching a distance of $b=10$ links from the global network. The probability of overtaking the global network, and thus successfully attacking the blockchain, is $\rho(x)$, which satisfies the following boundary value problem (BVP),
\begin{gather}
\rho'' + r \rho' = 0, \quad x \in (0, b) \\
\rho(0) = 1, \quad \rho(b) = 0.
\end{gather}
